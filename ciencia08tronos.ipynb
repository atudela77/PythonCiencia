```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/game_of_thrones_deaths_collecti.csv')
df
```

Necesito saber los dato con Nan.
Si son datos string los ponemos como Desconocido
Si son datos numéricos los ponemos como 0

```python
df.shape
```

```python
df.info()
```

```python
df = df.fillna({'allegiance': 'Desconocido', 'killer': 'Desconocido', 'killers_house': 'Desconocido'})
df.tail(20)
```

```python
df.info()
```

Me gustaría saber el número de muertes totales y también saber el número de muertes que hubo en la primera temporada

```python
muertes_totales = df.shape[0]
muertes_temp1 = df[df.season == 1].shape[0]
print(f"Número de muertes totales: {muertes_totales}")
print(f"Número de muertes en la primera temporada: {muertes_temp1}")
```

Quiero saber el porcentaje de muertes de la temporada 1 respecto al conjunto de asesinatos de la serie completa.

```python
ratio_temp1 = round((muertes_temp1/muertes_totales) * 100, 2)
print(f"El porcentaje de muertes en la primera temporada respecto al total es de {ratio_temp1}%")
```

Muertes y porcentaje de la segunda y tercera temporada

```python
muertes_temp2 = df[df.season == 2].shape[0]
muertes_temp3 = df[df.season == 3].shape[0]
ratio_temp2 = round((muertes_temp2/muertes_totales)*100,2)
ratio_temp3 = round((muertes_temp3/muertes_totales)*100,2)
```

Lo metemos todo en un diccinario.

```python
data = {"Temporada": ["Temporada 1", "Temporada 2", "Temporada 3"],
        "Muertes": [muertes_temp1, muertes_temp2, muertes_temp3],
        "Porcentaje": [ratio_temp1, ratio_temp2, ratio_temp3]}
df_muertes = pd.DataFrame.from_dict(data)
df_muertes
```

Pintamos el  gráfico

```python
df_muertes.plot(kind="bar", x="Temporada", y="Porcentaje")
```

Agrupar el número de muertes por temporada

```python
grupo_season = df.groupby(by = 'season')['killer']
grupo_season.describe()
```

Agrupamos un DataFrame por asesino

```python
df_killer = df.groupby('killer').sum()
df_killer
```

Me gustaría visualizar el asesiono y el número de muertes

```python
df_killer = df.groupby('killer')['death_no'].agg('count')
df_killer
```

Mostrar un gráfico Pie con los 10 primeros asesinos

```python
df_primeros = df_killer.head(10)
df_primeros.groupby('killer').sum().plot(kind='pie', y='count')
```

2) Mostrar las muertes de Daenerys Targaryen en un DataFrame

4) Mostrar la evolución de nuestra asesina por temporada en un gráfico de barras

Listar los 10 asesinos más prolíficos

```python
df_asesinos = df.groupby('killer').count()
df_asesinos_ord = df_asesinos.sort_values('death_no', ascending=False).head(10)
df_asesinos_ord.plot(kind="bar", y='death_no', ylabel= "Número de muertes")
```
