```python
import pandas as pd
import numpy as np
from matplotlib import pyplot
```

```python
df = pd.read_csv('data/marvel_movies.csv')
df.head()
```

```python
df.groupby('year').mean('duracion')
```

```python
def beneficios(row):
    rec = row['recaudacion']
    pre = row['presupuesto']
    return rec - pre
df['Beneficio'] = df.apply(beneficios, axis=1)
df
```
