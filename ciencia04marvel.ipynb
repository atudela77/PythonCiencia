Sacar información variada del fichero de películas de Marvel

Importamos las librerías de pandas y numpy

```python
import pandas as pd
import numpy as np
```

Guardar los datos del fichero marvel_movies.csv en un DataFrame

```python
df = pd.read_csv('data/marvel_movies.csv')
df
```

Ver el nombre de las columnas

```python
df.columns
```

Ver los índices de las filas

```python
df.index
```

Mostrar información de la estructura del DataFrame

```python
df.info()
```

Ordenar películas por año de estreno

```python
df_ordenado = df.sort_values('year')
df_ordenado
```

Mostrar aquellas películas con presupuesto mayor de 200 millones ordenadas de mayor a menor presupuesto

```python
df_ironman = df[df['presupuesto'] > 200].sort_values('presupuesto', ascending=False)
df_ironman
```

Mostrar las 3 películas que más recaudaron

```python
df_top3 = df.sort_values('recaudacion', ascending=False).head(3)
df_top3
```

Mostrar la peor valoración

```python
df_peorvaloracion = df['valoracion'].min()
df_peorvaloracion
```

Mostrar la película con peor valoración

```python
df_peorvalorada = df.sort_values('valoracion', ascending=False).tail(1)
df_peorvalorada
```

Mostramos la información estadística de la tabla

```python
df.describe()
```

Mostrar la recaudación total, el presupuesto medio, la valoración máxima

```python
df['recaudacion'].sum()
```

```python
df['presupuesto'].mean()
```

```python
df['valoracion'].max()
```

Transponer el fichero

```python
df.T
```

Metemos unos nulos en algún dato

```python
df['valoracion'][2:4] = np.nan
df
```

Buscamos los Nan en las columnas

```python
df.isna().sum()
```

Cambiamos los valores Nan por la valoración 7.1

```python
df['valoracion'] = df['valoracion'].fillna(7.1)
df
```
